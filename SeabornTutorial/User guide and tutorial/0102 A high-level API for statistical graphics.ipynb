# Seaborn 소개
## A high-level API for statistical graphics

- relplot 함수는 다양한 통계적 관계를 시각화하도록 설계되었기 때문에 

```python
# 라이브러리 불러오기
import seaborn as sns

# 기본 테마 적용
sns.set_theme()
```

```python
dots = sns.load_dataset('dots')

sns.relplot(
    data = dots, kind = 'line',
    x = 'time', y = 'firing_rate', col = 'align',
    hue = 'choice', size = 'coherence', style = 'choice',
    facet_kws = dict(sharex = False)
);
# x 축 : time
# y 축 : firing_rate
# 표 나누기 col : align
# 선 모양 기준 : choice
# 색 기준 : choice
# 선 크기 기준 : coherence
```

분산형 : 마커 영역과 기호를 변경하고 선 너비와 선 모양이 변경되어있음

```python
dots
```

## 통계적 추정
Seaborn 기능은 자동으로 필요한 통계적 추정을 수행한다.

```python
fmri = sns.load_dataset('fmri')
sns.relplot(
    data = fmri, kind = 'line',
    x = 'timepoint', y = 'signal', col = 'region',
    hue = 'event', style = 'event'
);
# x 축 : timepoint
# y 축 : signal
# 표 나눌 기준 : region
# hue 색기준 : event
# 선 style 기준 : event
```

부트스트랩을 사용하여 신뢰구간을 계산하고 추정치의 불확실성을 나타내는 errorbar를 그린다

```python
tips = sns.load_dataset('tips')
sns.lmplot(data = tips, x = 'total_bill', y = 'tip', col = 'time', hue = 'smoker');
```

lmport()로 선형 회귀 모델을 포함한 산점도를 표현할 수 있다.



## Distributional representations
- 통계 분석에는 데이터 세트의 변수 분포에 대한 지식이 필요하다.
- displot()는 분포시각화를 도움

```python
sns.displot(data = tips, x = 'total_bill', col = 'time', kde = True);
```

- displot(kind = 'ecdf') 이면 누적 분포기능을 보여줌
- kde : kernel density
https://seaborn.pydata.org/generated/seaborn.displot.html

```python
sns.displot(data = tips, kind = 'ecdf', x = 'total_bill', col = 'time', hue = 'smoker', rug = True);
```

## Plots for categorical data
- catplot는 분류데이터를 시각화
- 다양한 수준의 세분성을 제공

```python
sns.catplot(data = tips, kind = 'swarm', x = 'day', y = 'total_bill', hue = 'smoker');
```

```python
sns.catplot(data = tips, kind = 'violin', x = 'day', y = 'total_bill', hue = 'smoker', split = True);
```

```python
# 각 중첩된 범주 내에서 평균값과 신뢰구간만 보여줄 수 있다.
sns.catplot(data = tips, kind = 'bar', x = 'day', y = 'total_bill', hue = 'smoker');
```

```python
sns.catplot(data = tips, kind = 'bar', x = 'day', y = 'total_bill', hue = 'smoker', errorbar = None);
# 신뢰구간 삭제하려면 errorbar = None 해주면됨
```
