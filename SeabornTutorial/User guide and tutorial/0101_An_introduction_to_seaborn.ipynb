# Seaborn 소개
## Seaborn?
(오역, 번역기 말투 주의!!!)
Seaborn은 파이썬 그래프를 만들기 위한 라이브러리
matplotlib 으로 구축되며 판다스 데이터 구조와 가깝게 통합된다.

```python
# 라이브러리 불러오기
import seaborn as sns

# 기본 테마 적용
sns.set_theme()

# 샘플 데이터 불러오기
tips = sns.load_dataset('tips')

# 시각화
sns.relplot(
    data = tips, x = 'total_bill', y = 'tip', col = 'time',
    hue = 'smoker', style = 'smoker', size = 'size'
);
```

### 그래프 figure 하기
주피터나 ipython이 아닌 경우에 `matplotlib.pyplot.show()`를 불러와야한다.

### 테마 적용
- seaborn으로 만들지 않더라도 모든 matplotlib plot에 영향을 미친다.
- 기본테마 외에도 다른 것도 있음
- matplotlib 기본값을 사용하려면 이 단계를 뛰어 넘어가도 됨
